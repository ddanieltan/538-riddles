```python
import pandas as pd
import numpy as np
import re
```

```python
def is_abba(word):
    if word[0]!=word[1] and word[1]==word[2] and word[0]==word[3]:
        return True
    else:
        return False
```

```python
def check(word):
    if len(word)==4:
        if is_abba(word):
            return True
    elif len(word)>4:
        for i in xrange(len(word)-3):
            if is_abba(word[i:i+4]):
                return True
    return False
            
```

```python
# parse input into substrings outside [] and substrings inside []
# test all words inside brackets fail abba test
# test any word outside brackets pass abba test
counter=0
for t in test_inputs:
    all_words=filter(None,re.findall('\w+',t))
    in_brackets=filter(None,re.findall('\[(\w+)\]',t))
    out_brackets = [x for x in all_words if x not in in_brackets]
    
    if all(check(i) for i in in_brackets)==False:
        if any(check(o) for o in out_brackets)==True:
            print t
            counter+=1
    
print counter
```
